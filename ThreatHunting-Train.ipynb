```python
import pandas as pd

df_DDoS = pd.read_csv('Dataset\Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
df_DDoS.head()
```

```python
import numpy as np

df_DDoS.replace([np.inf, -np.inf], np.nan, inplace=True)
df_DDoS.isnull().sum()
```

```python
df_DDoS.describe()
```

```python
df_DDoS.info()
```

```python
dropable_column = [
    'Flow ID',
    ' Source IP',
    ' Source Port',
    ' Destination IP',
    ' Timestamp'
]

df_trainable = df_DDoS.drop(columns=dropable_column)
df_trainable.info()
```

```python
df_trainable.head()
```

```python
from matplotlib import pyplot as plt
import seaborn as sb

plt.figure(figsize=(8, 6))
sb.countplot(data=df_trainable, x=" Label")
plt.title("Count Plot of DDoS", fontsize=16)
plt.xlabel("DDoS", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks([0, 1], labels=["DDoS", "BENIGN"])
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

laen = LabelEncoder()

df_trainable[' Label']= laen.fit_transform(df_trainable[' Label'])

df_trainable.head()
```

```python
df_trainable.info()
```

```python
from sklearn.model_selection import train_test_split

target = [' Label']
x = df_trainable.drop(columns=target)
y = df_trainable[target]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import time

rf = RandomForestClassifier(n_estimators=100, random_state=42)
print("Begine training")
rf.fit(x_train, y_train.values.ravel())
print("Done training")
y_pred = rf.predict(x_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(8, 6))
sb.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for DDoS Detection', fontsize=16)   
plt.xlabel('Predicted Label', fontsize=14)
plt.ylabel('True Label', fontsize=14)
plt.xticks([0.5, 1.5], labels=["DDoS", "BENIGN"])
plt.yticks([0.5, 1.5], labels=["DDoS", "BENIGN"])
plt.show()
```

```python
impt_features = pd.Series(rf.feature_importances_, index=x_train.columns).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
impt_features.nlargest(10).plot(kind='barh', color='red')
plt.title('Top 10 Important Features for DDoS Detection', fontsize=16)
plt.show()
```
