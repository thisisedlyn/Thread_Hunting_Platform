```python
#open dataset
import pandas as pd

df_DDoS = pd.read_csv('Dataset\Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
df_DDoS.head()

#kalo pandas nya error bisa install dulu
#pip install pandas
#semua library sama kalo ga, bisa diinstall pake cara yang sama
```

Check for missing value

```python
import numpy as np

df_DDoS.replace([np.inf, -np.inf], np.nan, inplace=True)
df_DDoS.isnull().sum()
```

```python
#kalo nemu data yang aneh, benerin dulu, contoh yang di flow duration kan ga mungkin negatif, emagn ada waktu sebelum 0?
print("Jumlah data minus :", len(df_DDoS[df_DDoS[' Flow Duration'] < 0]))

df_DDoS = df_DDoS[df_DDoS[' Flow Duration'] >= 0]
print("After cleaning:", df_DDoS[' Flow Duration'].min())
```

```python
#cek kolom lain juga, saiapa tau ada yang minus

colmn = df_DDoS.select_dtypes(include=[np.number]).columns

min_colmn = [col for col in colmn if (df_DDoS[col] < 0).any()]

if len(min_colmn) > 0:
    for col in min_colmn:
        min_sum = (df_DDoS[col] < 0).sum()
        print(f"Negative data in '{col}': {min_sum}")
```

```python
#buat flow & Fwd IAN Min bisa di hapus aja

print ("before cleaning, Flow IAT Min : ", len (df_DDoS[df_DDoS[' Flow IAT Min'] < 0]))
print ("before cleaning, Fwd IAT Min : ", len (df_DDoS[df_DDoS[' Fwd IAT Min'] < 0]))
df_DDoS = df_DDoS[df_DDoS[' Flow IAT Min'] >= 0]
df_DDoS = df_DDoS[df_DDoS[' Fwd IAT Min'] >= 0]

print ("after cleaning, Flow IAT Min : ", len (df_DDoS[df_DDoS[' Flow IAT Min'] < 0]))
print ("after cleaning, Fwd IAT Min : ", len (df_DDoS[df_DDoS[' Fwd IAT Min'] < 0]))
```

```python
#Buat yang win bytes jangan di hapus yaa, ini lebih masuk akal buat di ganti jadi 0 aja yang minus

wb_colmn = ['Init_Win_bytes_forward', ' Init_Win_bytes_backward']

for col in wb_colmn:
    df_DDoS.loc[df_DDoS[col] < 0, col] = 0

print("Negative data in 'Init_Win_bytes_forward':", len(df_DDoS[df_DDoS['Init_Win_bytes_forward'] < 0]))
print("Negative data in 'Init_Win_bytes_backward':", len(df_DDoS[df_DDoS[' Init_Win_bytes_backward'] < 0])) 
```

```python
df_DDoS[['Init_Win_bytes_forward', ' Init_Win_bytes_backward']].head()
```

```python
#Cari infinite value, biar ga crash pas di training
colmn = df_DDoS.select_dtypes(include=[np.number]).columns
inf_colmn = [col for col in colmn if np.isinf(df_DDoS[col]).any()]

if len(inf_colmn) > 0:
    for col in inf_colmn:
        inf_sum = np.isinf(df_DDoS[col]).sum()
        print(f"Infinite data in '{col}': {inf_sum}")
        
```

```python
df_DDoS[['Flow Bytes/s', ' Flow Packets/s']].head()
```

```python
df_DDoS.replace([np.inf, -np.inf], np.nan, inplace=True)
df_DDoS.dropna(inplace=True)
df_DDoS[['Flow Bytes/s', ' Flow Packets/s']].head()
```

```python
df_DDoS.info()
```

```python
#Nah ini tuh buat buang column yang kurang kepake buat trainig, Lebih bututh pola penyerangannya daripada IP si Cracker
dropable_column = [
    'Flow ID',
    ' Source IP',
    ' Source Port',
    ' Destination IP',
    ' Timestamp'
]

df_trainable = df_DDoS.drop(columns=dropable_column)
df_trainable.info()

"""
nah kalo udah kehapus dan udah bersih bisa di balikin ke csv lagi supaya bisa di training
pake code ini :
df_trainable.to_csv('Dataset\Friday-WorkingHours-Afternoon-DDos_Cleaned.csv', index=False)
"""
```

```python
df_trainable.head()
```

```python
from sklearn.preprocessing import LabelEncoder

laen = LabelEncoder()

df_trainable[' Label']= laen.fit_transform(df_trainable[' Label'])

df_trainable.head()
```

```python
df_trainable.info()
```

```python
from sklearn.model_selection import train_test_split

target = [' Label']
x = df_trainable.drop(columns=target)
y = df_trainable[target]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from matplotlib import pyplot as plt
import seaborn as sb

plt.figure(figsize=(8, 6))
sb.countplot(data=y_train, x=" Label")
plt.title("Count Plot of DDoS Before Split", fontsize=16)
plt.xlabel("DDoS", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks([0, 1], labels=["DDoS", "BENIGN"])
plt.show()


y_train.value_counts(normalize=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import time

rf = RandomForestClassifier(n_estimators=100,max_depth=10, random_state=42)
print("Begine training")
rf.fit(x_train, y_train.values.ravel())
print("Done training")
y_pred = rf.predict(x_test)

print("Classification Report:")
print(classification_report(y_test, y_pred, digits=5))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
plt.figure(figsize=(8, 6))
sb.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for DDoS Detection', fontsize=16)   
plt.xlabel('Predicted Label', fontsize=14)
plt.ylabel('True Label', fontsize=14)
plt.xticks([0.5, 1.5], labels=["DDoS", "BENIGN"])
plt.yticks([0.5, 1.5], labels=["DDoS", "BENIGN"])
plt.show()
```

```python
impt_features = pd.Series(rf.feature_importances_, index=x_train.columns).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
impt_features.nlargest(10).plot(kind='barh', color='red')
plt.title('Top 10 Important Features for DDoS Detection', fontsize=16)
plt.show()
```
