```python
file_path = 'Dataset\Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv'
df_webAttack = pd.read_csv(file_path, encoding='latin1' )

if ' Label' in df_webAttack.columns:
    df_webAttack[' Label'] = df_webAttack[' Label'].str.replace('\x96', '-')
    df_webAttack=df_webAttack.dropna(subset=[' Label'])
```

```python
df_webAttack[' Label'].unique()
```

```python
rename = {
    'Web Attack - Brute Force': 'Brute Force',
}

df_webAttack[' Label'] = df_webAttack[' Label'].replace(rename)

df_webAttack[' Label'].unique()
```

```python
print("Each Value : ", df_webAttack[' Label'].value_counts())
print("sum = ", df_webAttack[' Label'].value_counts().sum())
```

Benign Data

```python
df_benign = pd.read_csv('Dataset\Monday-WorkingHours.pcap_ISCX.csv')
df_benign[' Label'].value_counts()
```

Data Comparation

```python
benign = df_portscan[df_portscan[' Label'] == 'BENIGN'].copy()
ddos = df_DDoS[df_DDoS[' Label'] == 'DDoS'].copy()
portScan = df_portscan[df_portscan[' Label'] == 'PortScan'].copy()
bruteForce = df_bruteForce[df_bruteForce[' Label'] == 'Brute Force'].copy()
burtforce2 = df_webAttack[df_webAttack[' Label'] == 'Brute Force'].copy()

df_combined = pd.concat([benign, ddos, portScan, bruteForce, burtforce2], ignore_index=True)
df_combined[' Label'].unique()
```

```python
df_combined.head()
```

```python
df_combined[' Label'].unique()
```

```python
from matplotlib import pyplot as plt
import seaborn as sb

plt.figure(figsize=(18, 6))
sb.countplot(data=df_combined, x=" Label")
plt.title("Data Distribution", fontsize=16)
plt.xlabel("Attack Type", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks([0, 1, 2, 3], labels=["BENIGN", "DDoS", "PortScan", "Brute Force"])
plt.show()

df_combined[' Label'].value_counts()
```
