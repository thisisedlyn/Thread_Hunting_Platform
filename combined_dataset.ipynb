```python
import pandas as pd

df_DDoS = pd.read_csv('Dataset\Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
df_DDoS.info()
```

```python
df_portscan[' Label'].unique()
```

```python
df_bruteForce = pd.read_csv('Dataset\Tuesday-WorkingHours.pcap_ISCX.csv')
df_bruteForce.info()
```

```python
df_bruteForce[' Label'].unique()
```

```python
rename = {
    'FTP-Patator': 'bruteForce-FTP',
    'SSH-Patator': 'bruteForce-SSH',
}

df_bruteForce[' Label'] = df_bruteForce[' Label'].replace(rename)

df_bruteForce[' Label'].unique()
```

```python
file_path = 'Dataset\Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv'
df_webAttack = pd.read_csv(file_path, encoding='latin1' )

if ' Label' in df_webAttack.columns:
    df_webAttack[' Label'] = df_webAttack[' Label'].str.replace('\x96', '-')
    df_webAttack=df_webAttack.dropna(subset=[' Label'])
```

```python
df_webAttack[' Label'].unique()
```

```python
files = [df_DDoS, df_portscan, df_bruteForce, df_webAttack]

df_combined = pd.concat(files, ignore_index=True)
df_combined.head()
```

```python
df_combined[' Label'].unique()
```

```python
df_combined.to_csv('Dataset/finalThreat-Dataset.csv', index=False)
```
